### Q1: What country has emitted the most per capita CO2 in 2021 using one fuel type?

```python
emissions_trim['Year'].max()
```

```python
df_2021 = emissions_trim.query("Year == 2021")
df_2021
```

```python
max_2021 = df_2021['emissions'].max()
max_2021
```

```python
most_2021 = df_2021.query("emissions == @max_2021")
most_2021
```

```python
print(f"The country that produced the most per capita CO2 emissions from a single fuel type in 2021 was Qatar.")
```

### Q2: What country emitted the least CO2 across all years documented?

```python
country_totals = emissions_trim.groupby("Country", as_index=False)["emissions"].sum()
```

```python
country_totals['emissions'].min()
```

```python
country_totals.query("emissions == 0.0")
```

### Q3: How much CO2 emission per capita was documented across all countries per year?

```python
year_totals = emissions_trim.groupby("Year", as_index=False)["emissions"].sum()
year_totals
```

### Q4: What year had the most CO2 emissions per capita across all countries and fuel types?

```python
year_totals['emissions'].max()
```

```python
year_totals.query("emissions == 1373.15072")
```

```python
print(f"1956 had the most emissions per capita, according to the dataset.")
```

### Q6: What fuel type produced the most CO2 emissions per capita?

```python
fuel_totals = emissions_trim.groupby("fuel", as_index=False)["emissions"].sum()
fuel_totals
```
