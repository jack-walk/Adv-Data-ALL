```python
import pandas as pd
```

```python
## Importing CSV file of all 311 service requests over tree overgrowth since 2020.

df = pd.read_csv("311_Service_Requests_from_2020_to_Present_20260418.csv", low_memory=False)
df
```

```python
## Filtering out entries prior to 2023 for time relevance.

df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')
dfRecent = df[df['Created Date'] >= pd.Timestamp('2023-01-01 00:00:00')]
dfRecent
```

```python
## Examining all possible zip codes.

zip_codes_all = dfRecent['Incident Zip'].unique()
print(zip_codes_all)
```

```python
## Counting reports per zip code.

zip_counts = (
    dfRecent['Incident Zip']
    .value_counts(dropna=False)
    .reset_index()
)

zip_counts.columns = ['Incident Zip', 'Count']
zip_counts.head(15)
```

```python
## Returning dataframe as a CSV to visualize in Datawrapper

zip_counts.to_csv("zip_counts.csv", index=False)
```
